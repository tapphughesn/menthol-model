```python
# Average transition probabilites (weighted)
print("Average Transition Probabilities in 2018-2019 With and Without Initiation/Continuation Calibration")

all_probs = np.concatenate([probs1, probs2345], axis=0)
all_probsc = np.concatenate([probs1c, probs2345c], axis=0)

weights = np.concatenate([ 
    arr1[:,15],
    arr2345[:,15]
])
weightsc = np.concatenate([ 
    arr1c[:,15],
    arr2345c[:,15]
])

avg_probs = np.sum(all_probs * weights[:, np.newaxis], axis=0) / np.sum(weights)
avg_probsc = np.sum(all_probsc * weightsc[:, np.newaxis], axis=0) / np.sum(weightsc)

header = ["Initiation/Continuation\nCalibration?", "Never Smoker", "Former Smoker", "Menthol\nCigarette User", "Non-menthol\nCigarette User", "E-cigarette/Dual\nUser"]
row1 = ["No"] + [f"{np.around(i * 100, decimals=2)}%" for i in avg_probs]
row2 = ["5.5% Reduction"] + [f"{np.around(i * 100, decimals=2)}%" for i in avg_probsc]
printtab([header, row1, row2])
```
