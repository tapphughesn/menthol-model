```python
import numpy as np
from tabulate import tabulate
printtab = lambda x : print(tabulate(x, headers='firstrow'))
```

```python
print("methol smoking across years in total population")
years = [2016,2017,2018,2019,2020]
props = []
for year in years:
    arr2345 = np.load(f"/Users/nick/Documents/Gillings_work/outputs/2016-20_analysis/{year}arr2345calibpre.npy")

    arr2345 = arr2345[arr2345[:,11] <= 64]

    menthol_smokers = np.sum(arr2345[arr2345[:,6] == 1, 15])
    nonmenthol_smokers = np.sum(arr2345[arr2345[:,7] == 1, 15])

    prop = menthol_smokers / (menthol_smokers + nonmenthol_smokers)
    props.append(prop)

printtab([years, [f"{np.around(i * 100, decimals=2)}%" for i in props]])
```
