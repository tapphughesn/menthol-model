```python
"""Compare one menthol ban scenario to one status quo scenario"""
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
output_status_quo = np.load(
    "/Users/nick/Documents/Gillings_work/outputs/numpy_arrays/np_output_calibrated_2022-10-31_12-51-06-159823.npy"
)

output_menthol_ban = np.load(
    "/Users/nick/Documents/Gillings_work/outputs/numpy_arrays/np_output_calibrated_menthol_ban_1_5_2022-10-31_13-53-18-609609.npy"
)
```

```python
output_sq = np.sum(output_status_quo, axis=(1,2))
output_mb = np.sum(output_menthol_ban, axis=(1,2))

output_sq = output_sq[:,:-1]
output_mb = output_mb[:,:-1]

menthol_over_all_sq = output_sq[:,2] / np.sum(output_sq, axis=1)
menthol_over_all_mb = output_mb[:,2] / np.sum(output_mb, axis=1)

nonmenthol_over_all_sq = output_sq[:,3] / np.sum(output_sq, axis=1)
nonmenthol_over_all_mb = output_mb[:,3] / np.sum(output_mb, axis=1)

smoker_over_all_sq = (output_sq[:,2] + output_sq[:,3]) / np.sum(output_sq, axis=1)
smoker_over_all_mb = (output_mb[:,2] + output_mb[:,3]) / np.sum(output_mb, axis=1)



```

```python
fig, ax = plt.subplots(1,1,figsize=(9,4),dpi=150)
start = 2016
end = 2067
x0 = np.arange(start,end)
ax.set_xticks(np.arange(start, end, 5))
ax.plot(x0, menthol_over_all_sq * 100, c='b')
ax.plot(x0, menthol_over_all_mb * 100, c='g')
ax.scatter(x0, menthol_over_all_sq  * 100, c='b', s=15)
ax.scatter(x0, menthol_over_all_mb  * 100, c='g', s=15)
ax.legend(labels=[
    f"status quo scenario", 
    f"menthol ban scenario", 
])
plt.title("Comparison of Menthol Smoking Rate\nin Status Quo and Menthol Ban Scenarios")
ax.axvline(2021, ymin=0.1, ymax=0.99, color='r', linestyle='dashed')
ax.text(2022, 6, "Menthol ban start", c='r')
ax.set_ylabel("Menthol Smoking Percentage")
```
