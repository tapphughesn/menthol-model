```python
"""Seeing how to sample from Dirichlet and what the results might look like."""
```

```python
import numpy as np
from tabulate import tabulate
```

```python
printtab = lambda x : print(tabulate(x, headers='firstrow'))
```

```python
alpha = np.array([0.24, 0.44, 0.19, 0.14])*1000
s = np.random.dirichlet(alpha, 10000)
```

```python
s
```

```python
header = [
    "quit",
    "swtich to nonmenthol",
    "continue using menthol",
    "switch to ecig/dual",
]
tab = [header] + [[f"{np.around(100* e, 1)}%" for e in row] for row in s[:10]]
```

```python
printtab(tab)
```

```python
print(np.mean(s, axis=0))
print(np.std(s, axis=0))
```

```python
print(np.min(s, axis=0))
print(np.max(s, axis=0))
```

```python
p = np.array([[0.24, 0.44, 0.19, 0.14]] * 10)
```

```python
k = p + 0.1*(s[:10] - 0.25)
```

```python

header = [
    "quit",
    "swtich to nonmenthol",
    "continue using menthol",
    "switch to ecig/dual",
]
tab = [header] + [[f"{np.around(100* e, 1)}%" for e in row] for row in k]
printtab(tab)
```
