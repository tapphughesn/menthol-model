```python
probs1 = np.random.rand(10)
probs2 = np.random.rand(10)

probs = [probs1, probs2, 2 - probs1 - probs2]

probs = np.asarray(probs).transpose() / 2

probs
```

```python
chance = np.random.rand(probs.shape[0])[:,np.newaxis]
chance
```

```python
forward234 = np.concatenate([chance < np.sum(probs[:,:i], axis=1)[:,np.newaxis] for i in range(1, probs.shape[1] + 1)], axis=1)
backward234 = np.concatenate([(1-chance) < np.sum(probs[:,i:], axis=1)[:,np.newaxis] for i in range(probs.shape[1])], axis=1)

new_states234_uncut = forward234 * backward234 # need to do a AND

forward234, backward234
```
